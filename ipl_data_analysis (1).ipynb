```python
import matplotlib.pyplot as plt
matches['WinningTeam'].value_counts().head(10).plot(kind='bar',figsize=(10,5))
plt.title("top 10  match with wins")
plt.xlabel("team")
plt.ylabel("wins")
plt.xticks(rotation=45)
plt.show()
```

# toss winner distribution


```python
matches['TossWinner'].value_counts().head(10).plot(kind='bar',figsize=(10,5))
plt.title("top toss winning teams")
plt.xlabel("team")
plt.ylabel("toss wins")
plt.xticks(rotation=45)
plt.show()    

    
```

# player of the match(top 10)

```python
matches['Player_of_Match'].value_counts().head(10).plot(kind='bar',figsize=(10,5))
plt.title("top 10 player of the match")
plt.xlabel("player")
plt.ylabel("awards")
plt.xticks(rotation=45)
plt.show()
```

# top 10 run scorer

```python
top_batsmen=balls.groupby('batter')['batsman_run'].sum().sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar',figsize=(10,5))
plt.title("top 10 run scoreres")
plt.xlabel("batsman")
plt.ylabel("runs")
plt.xticks(rotation=45)
plt.show()
```

# top 10 wicket takers

```python
top_bowlers=balls[balls['isWicketDelivery']==1]['bowler'].value_counts().head(10)
top_bowlers.plot(kind='bar',figsize=(10,5))
plt.title("top 10 wicket takers")
plt.xlabel("bowlers")
plt.ylabel("wickets")
plt.xticks(rotation=45)
plt.show()
```

# matches per seasson 

```python
matches['Season'].value_counts().sort_index().plot(kind='line',marker='o',figsize=(10,5))
plt.title("matches played per season")
plt.xlabel("season")
plt.ylabel("number of matches")
plt.show()
```

# toss decision analysis

```python
matches['TossDecision'].value_counts().plot(kind='pie',autopct='%1.1f%%',figsize=(6,6))
plt.title("toss decision")
plt.ylabel("")
plt.show()
```

# top 10 venues

```python
matches['Venue'].value_counts().head(10).plot(kind='bar',figsize=(10,5))
plt.title("top 10 venues by number of matches")
plt.xlabel("Venues")
plt.ylabel("matches")
plt.xticks(rotation=90)
plt.show()
```

# matches wonby runs vs wicket

```python
matches['WonBy'].value_counts().plot(kind='bar')
plt.title("matches wonbr runs vs wicket")
plt.xlabel("won by")
plt.ylabel("count")
plt.show()
```
